```python
from dotenv import load_dotenv
load_dotenv()

import os
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import json
```

```python
def get_metadata(data_file):
    with open(data_file, 'r') as f:
        for line in f:
            yield line
```

### Load data from json

```python
PATH = os.getenv('INPUT')

data_file = PATH
```

```python
metadata = get_metadata(PATH)

paper_dict = []
```

```python
# Don't run this in Jupyter, your computer is gonna crash (unless your comp is too strong)
for paper in metadata:
    paper_dict.append(json.loads(paper))
```

```python
author_dict = {}
```

### Get the data of authors

```python
for i in range(len(paper_dict)):
    for e in paper_dict[i]['authors_parsed']:
        e = ' '.join(e[::-1])
        if e in author_dict:
            author_dict[e] = author_dict[e] + 1
        else:
            author_dict[e] = 1
```

```python
print(len(author_dict))
```

```python
author_dict = sorted(author_dict.items(), key = lambda x: x[1], reverse=True)
```

### Saving data to json file

```python
json_file = json.dumps(author_dict)

OUTPUT_PATH = os.getenv('OUTPUT')
f = open(OUTPUT_PATH, 'w')
f.write(json_file)
f.close()
```
